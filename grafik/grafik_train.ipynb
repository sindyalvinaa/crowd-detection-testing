```python
import matplotlib.pyplot as plt

# Data model dan akurasinya
model_names = ['YOLOv8n', 'YOLOv8s', 'YOLOv8m', 'YOLOv8l', 'YOLOv8x']
accuracies = [0.87, 0.90, 0.91, 0.91, 0.92]

# Membuat diagram garis
plt.figure(figsize=(10, 6))
plt.plot(model_names, accuracies, marker='o', linestyle='-', color='skyblue')

# Menambahkan judul dan label
plt.title('Accuracy Comparation of YOLOv8 Models')
plt.xlabel('Models')
plt.ylabel('Accuracy')

# Menampilkan nilai akurasi di atas setiap titik
for i in range(len(model_names)):
    plt.text(i, accuracies[i] + 0.001, str(accuracies[i]), ha='center')

# Menampilkan grafik
plt.ylim(0.70, 1.00)
plt.grid(True)
plt.show()
```
